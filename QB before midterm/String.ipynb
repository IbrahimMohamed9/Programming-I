```python
"""Task #1"""

def check(str1, str2):
    x = ""
    z = ""
    f = ""
    for i in str1:
        if i.isalpha and i not in x:
            x += i
    for j in str2:
        for k in x:
            if k == j:
                z += k
    for _ in z:
        if _ not in f:
            c = list(z).count(_)
            if c > 1:
                f += _
    return f


word1 = "aaddffgghhjjkklzzl"
word2 = "asdfghjklz"
check(word2, word1)
```

```python
"""Task #4"""

def countTriple(n):
    m = 0
    for i in n:
        count = 0
        k = ""
        k += i
        for i in n:
            if i == k:
                count += 1
                if count >= 3:
                    m += 1
    return m // 3


print(countTriple("abcXXXabc"))
print(countTriple("xxxabyyyycd"))
print(countTriple("a"))
```

```python
"""Task #11"""

def make_abba(a, b):
    z = ""
    n = 1
    while n < 5:
        if n == 1:
            z += a
            n += 1
        elif n == 2:
            z += b
            n += 1
        elif n == 3:
            z += b
            n += 1
        else:
            z += a
            n += 1
    return z


print(make_abba('Hi', 'Bye'))
print(make_abba('Yo', 'Alice'))
print(make_abba('What', 'Up'))
```

```python
"""Task #17"""

def combo_string(n, n1):
    if len(n) > len(n1):
        return n1 + n + n1
    elif len(n1) > len(n):
        return n + n1 + n
    else:
        return "Those have same length"


print(combo_string('Hello', 'hi'))
print(combo_string('hi', 'Hello'))
print(combo_string('aaa', 'b'))
```

```python
"""Task #21"""

def count_hi(n):
    count = 0
    for i in range(2, len(n) + 1):
        z = n[(i - 2):i]
        if z == "hi":
            count += 1
    return count


print(count_hi('abc hi ho'))
print(count_hi('ABChi hi'))
print(count_hi('hihi'))
```

```python
"""Task #22"""

def cat_dog(n):
    c = "cat"
    d = "dog"
    countd, countc = 0, 0
    for i in range(len(c), len(n) + 1):
        z = n[(i - len(c)):i]
        if z == c:
            countc += 1
    for i in range(len(d), len(n) + 1):
        a = n[(i - len(d)):i]
        if a == d:
            countd += 1
    if countd == countc:
        return True
    else:
        return False


print(cat_dog('catdog'))
print(cat_dog('catcat'))
print(cat_dog('1cat1cadodog'))
```

```python
"""Task #24"""

def end_other(str1, str2):
    str1 = str1.lower()
    str2 = str2.lower()
    if len(str1) > len(str2):
        if str1[-len(str2):] == str2:
            return True
    elif len(str2) > len(str1):
        if str2[-len(str1):] == str1:
            return True
    elif str2 == str1:
        return True
    else:
        return False
    

print(end_other('Hiabc', 'abc'))
print(end_other('AbC', 'HiaBc'))
print(end_other('abc', 'abXabc'))
```

```python
"""Task #30"""

def non_repeating(n):
    n1 = len(n) - 1
    for i in n:
        x = ""
        for j in n:
            if i != j:
                x += i
            x1 = len(x)
            if x1 == n1:
                return i

            
print(non_repeating("hhhaaaaalloooowwwwwzbb"))
```

```python
"""Task #33"""

def max_consecutive_0(n):
    x = 0
    z = n.split("1")
    for i in z:
        if len(i) > x:
            x = len(i)
    return x
    

print(max_consecutive_0('111000010000110')) # 4
print(max_consecutive_0('111000111')) # 3
```

```python
"""Task #34"""

def replace_words(n):
    z = n.split(" ")
    q = z[-1]
    p = ""
    for i in z:
        if len(i) >= 5 and i != q:
            p += (("#" * len(i))+" ")
        elif q == i:
            if len(i) >= 5:
                p += ("#" * len(i))
            else:
                p += i
        else:
            p += (i + " ")  
    return p

replace_words('Count the lowercase letters in the said list of words')
# """
# ##### the ######### ####### in the said list of ######
# """
```
